# convexhull package demonstration

Within this notebook, we see how we use the $\texttt{convexhull}$ R package to obtain the convex hull of a set of points.

Firstly, we need to install and use the library $\texttt{devtools}$, in order to import the $\texttt{convexhull}$ R package.

```r
install.packages("devtools")
# loading the library
library(devtools)
```

Next we install the $\texttt{convexhull}$ R package from the rj-lambert/convexhull github repository.

```r
install_github("rj-lambert/convexhull")
# loading the library
library(convexhull)
```

Following this, we can use the functions within the package to obtain the convex hull of a set of points, by implementing the Jarvis March algorithm.

Within the package, the function $\texttt{jarvis_march()}$ will output the line segments that form the polygon whose vertices are the convex hull. This function can take both a list of length 2 vectors or a 2-column dataframe as inputs.

```r
# implementing the jarvis_march function with a list of vectors

# creating a list of length 2 vectors
testlist = list(c(1,1), c(2,2), c(3,3), c(4,4))

jarvis_march(testlist)
```

```r
# implementing the jarvis_march function with a dataframe

# creating a dataframe of x,y pairs
x = rnorm(20,0,1)
y = rnorm(20,0,1)

df = data.frame(x,y)

jarvis_march(df)
```

In addition to outputting the line segments, there is a function $\texttt{jarvis_plot()}$, which, given similar imput to $\texttt{jarvis_march()}$ will plot the input points in black, the convex hull in blue, with the connecting line segments in red. Again, this works with both a list of vectors or a dataframe as input.

```r
# using jarvis_plot with a list of vectors

jarvis_plot(testlist)
```

```r
# using jarvis_plot with a dataframe

jarvis_plot(df)
```

The third function included in the package is the function $\texttt{in_hull()}$, which, given a single point, as well as an initial set of points, will determine whether the single point is part of the set of points that form the convex hull. In addition to this, it produces a plot similar to $\texttt{jarvis_plot()}$, with the addition of the point of interest, highlighted in green. Again, this works with both a list of vectors and a dataframe as input.

```r
# using in_hull with a list of vectors

in_hull(c(1,2), testlist)
```

```r
# using in_hull with a list of vectors

in_hull(c(1,1), testlist)
```

```r
# using in_hull with a dataframe

for (i in 1:nrow(df)){
  in_hull(df[i,], df)
}
```
